```python
pd.get_option("display.max_columns")
df = pd.read_csv("taishi.csv")
sevens = ["加藤清正", "福島正則", "加藤嘉明", "平野長泰",  "脇坂安治", "糟屋武則", "片桐且元"]

def busho_finder(df, list):
    index_list=[]
    for idx1, name in enumerate(list):
        for idx2, row in df.iterrows():
            if ((row["武将姓"] == name[:2]) & (row["武将名"] == name[2:4])):
                index_list.append(idx2)
    return index_list

# busho_finder(df, sevens)
df = df.iloc[busho_finder(df, sevens)]
df
```

```python
plt.scatter("統率", "知略", s=100, alpha=1, data=df)
plt.rcParams['figure.figsize'] = (12.0, 8.0)
plt.rcParams["font.size"] = 15
plt.title("賤ヶ岳7本槍の能力値　散布図")
plt.xlabel("統率")
plt.ylabel("知略")
for idx, row in df.iterrows():
    plt.annotate(row["武将姓"]+row["武将名"], (row["統率"], row["知略"]))
# plt.savefig("seven.png")
```

```python
# deep, muted, bright, pastel, dark, colorblind
sns.set(context="notebook", style="ticks", palette="bright")
df= pd.read_csv("dataset/tmdb_5000_movies.csv")
df = df[~df["release_date"].isna()]
df["decade"] = df["release_date"].apply(lambda x: x[:3] + "0s")
df["evaluation"] = df["vote_average"].apply(lambda x: "high" if x >= df["vote_average"].median() else "low")
```

```python
df["evaluation"].value_counts()
```

```python
sns.violinplot(x="decade", y="popularity", data=df.sort_values("decade"), hue="evaluation", palette="twilight")
```

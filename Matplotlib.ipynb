```python
taishi_df = pd.read_csv("taishi.csv")
taishi_count = taishi_df.groupby("武将姓").count()["武将名"]
taishi_count = taishi_count[taishi_count.apply(lambda x:x>5)]
```

```python
taishi_index = taishi_count.index
```

```python
count_df = taishi_df.groupby("武将姓").count().loc[taishi_index]["武将名"]
count_df
```

```python
taishi_df = taishi_df.groupby("武将姓").mean().loc[taishi_index][["戦闘", "政治"]]
```

```python
taishi_df = taishi_df.merge(count_df, left_on="武将姓", right_on="武将姓")
taishi_df = taishi_df.rename(columns={'武将名': '武将数'})
# taishi_df.index
# taishi_df["戦闘"].loc(taishi_df.index[0])
taishi_df["武将数"] = taishi_df["武将数"].apply(lambda x:x*50)
```

```python
c = np.arange(len(taishi_df))
plt.scatter("戦闘", "政治", s="武将数", c=c, alpha=0.5, data=taishi_df)
plt.rcParams['figure.figsize'] = (34.0, 14.0)
# plt.figure(figsize=(3, 1))
plt.rcParams["font.size"] = 20
plt.title("信長の野望大志　パラメーターで見る各家の優秀度")
plt.xlabel("戦闘パラメーター合計の平均値")
plt.ylabel("政治パラメーター合計の平均値")
for name in taishi_df.index:
    plt.annotate(name, (taishi_df["戦闘"].loc[name],taishi_df["政治"].loc[name]))
plt.savefig("taishi_param.png")
from matplotlib.backends.backend_pdf import PdfPages

pdf = PdfPages("taishi_param.pdf")

fig, axes = plt.subplots()
axes.plot(x, x**
pdf.savefig(fig)
pdf.close()
```

```python
df = pd.read_csv("dataset/tmdb_5000_movies.csv")
plt.rcParams['figure.figsize'] = (6.0, 4.0)
plt.rcParams["font.size"] = 10
plt.hist(df["vote_average"], bins=50)
plt.show()
```

```python
df = df[(df["revenue"] != 0) & (df["budget"] != 0)]
```

```python
df["profit"] = df.apply(lambda x: x["revenue"]-x["budget"], axis=1)
```

```python
fig, axes = plt.subplots()
logbins = np.logspace(5, np.log10(df["profit"].max()), 100)
axes.hist(df["profit"], bins=logbins)
axes.set_xscale("log")
plt.show()
```

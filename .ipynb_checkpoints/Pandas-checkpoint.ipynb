```python
df = pd.read_csv("dataset/tmdb_5000_movies.csv")
df.head(1)
```

```python
# def return_nan_if_zero(budget):
# #     if budget == 0:
# #         return np.nan
# #     else:
# #         return budget
   
```

```python
df["budget"] = df["budget"].apply(lambda budget: np.nan if budget==0 else budget)
df["revenue"] = df["revenue"].apply(lambda revenue: np.nan if budget==0 else revenue)
```

```python
df.head(1)
```

```python
df = df.dropna(subset = ["budget","revenue"])
```

```python
df["profit"] = df.apply(lambda row:row["revenue"] - row["budget"], axis=1)
```

```python
df
df[df["profit"] < 0]["profit"].count()/len(df)
```
